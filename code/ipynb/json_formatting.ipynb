```python
import pandas as pd
```

```python
data = "../csv_data/clean_vgsales.csv"
games_df = pd.read_csv(data) 
games_df.head()
```

```python
games_df = games_df.head(100).copy()
```

```python
games_df.columns
```

```python
games_df = games_df.rename(columns = {'Rank':'rank', 'Name':'name', 'Platform':'platform', 
                            'Year':'year', 'Genre':'genre', 'Publisher':'publisher', 
                            'NA_Sales':'na_sales', 'EU_Sales':'eu_sales', 'JP_Sales':'jp_sales', 'Other_Sales':'other_sales','Global_Sales':'global_sales'}).copy()
games_df.head()
```

```python
games_df.to_json("../../app/static/data/games.json",orient ="index")
```

```python
games_sorted = games_df.sort_values(by=['year'])
games_sorted
```

```python
games_sorted.to_json("../../app/static/data/games_sorted.json", orient="index")
```
