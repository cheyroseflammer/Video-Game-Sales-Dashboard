## Importing libraries

```python
# Imports
from bs4 import BeautifulSoup
import requests
```

## Checking out the whole document to see what elements to grab

```python
# Url we will be scraping
url = "https://www.wepc.com/news/video-game-statistics/"
results = requests.get(url).text
doc = BeautifulSoup(results, "html.parser")
doc
```

## Grabbing the stats cards from website

```python
stats = doc.find_all('p', {"class": "elementor-blockquote__content"})
stats
```
